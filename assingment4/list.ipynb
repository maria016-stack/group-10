```python
subjects=["Math","Physics","Chemistry","Biology","CS"]
#To add a subject at the end of the list
subjects.append("History")
print(subjects)
#To add a subject at a certain specified position
subjects.insert(3,"Language")
print(subjects)
#To remove a subject from the list with the specified value
subjects.remove("Chemistry")
print(subjects)
#To remove a subject at a specified position
subjects.pop(3)
print(subjects)
#To arrange subjects in an alphabetical order from A to Z
subjects.sort()
print(subjects)
#To rearrange subjects in an alphabetical order from Z to A
subjects.reverse()
print(subjects)
```
